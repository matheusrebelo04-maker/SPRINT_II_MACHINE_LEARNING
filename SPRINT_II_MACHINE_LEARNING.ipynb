# **Importação das bibliotecas**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
```

# **Criação da base de dados**

```python
import pandas as pd
import numpy as np

np.random.seed(42)

dados = {
    'Complexidade': np.random.choice(
        ['Baixa', 'Media', 'Alta'],
        50
    ),

    'Equipe': np.random.randint(
        3,
        20,
        50
    ),

    'Orcamento': np.random.randint(
        50000,
        500000,
        50
    ),

    'Atraso_Dias': np.random.randint(
        0,
        90,
        50
    ),

    'Risco': np.random.choice(
        ['Baixo', 'Medio', 'Alto'],
        50
    )
}

df = pd.DataFrame(dados)

df.head()
```

# **Verificando informações da base**

```python
df.info()
```

# **Verificando valores nulos**

```python
df.isnull().sum()
```

# **Verificando dados duplicados**

```python
df.duplicated().sum()
```

# **Transformar variáveis categóricas em números**

```python
encoder = LabelEncoder()

df['Complexidade'] = encoder.fit_transform(df['Complexidade'])
df['Risco'] = encoder.fit_transform(df['Risco'])

df.head()
```

# **Estatísticas básicas**

```python
df.describe()
```

# **Gráfico de distribuição do risco**

```python
sns.countplot(x='Risco', data=df)

plt.title('Distribuição dos Riscos')
plt.show()
```

# **Gráfico de atraso por risco**

```python
sns.boxplot(
    x='Risco',
    y='Atraso_Dias',
    data=df
)

plt.title('Atraso x Risco')
plt.show()
```

# **Gráfico de equipe por risco**

```python
sns.boxplot(
    x='Risco',
    y='Equipe',
    data=df
)

plt.title('Equipe x Risco')
plt.show()
```

# **Gráfico de orçamento por atraso**

```python
sns.scatterplot(
    x='Orcamento',
    y='Atraso_Dias',
    hue='Risco',
    data=df
)

plt.title('Orçamento x Atraso')
plt.show()
```

# **Matriz de correlação**

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    df.corr(),
    annot=True
)

plt.title('Matriz de Correlação')
plt.show()
```

# **Insights iniciais**


• Projetos com maior atraso tendem a apresentar maior risco.

• Projetos com maior complexidade normalmente possuem maior orçamento.

• Equipes menores podem apresentar mais dificuldades durante a implantação.

• O atraso parece ser uma das variáveis mais importantes para prever o risco.
